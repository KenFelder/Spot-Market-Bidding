```python
import matplotlib.pyplot as plt
import numpy as np

r = np.linspace(-1, 1, 400)
tau = np.exp(-r**2)

plt.plot(r, tau, label='Example curve')
plt.annotate(r'$\theta = 2$', xy=(0.5, 0.5), xytext=(0.6, 0.5), arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.xlabel('Aggressiveness Level, r')
plt.ylabel('Target Price, τ')
plt.title("Aggressiveness Model")
plt.legend()
plt.show()
```

```python
# plot the same graph with a dotted line
plt.plot(r, tau, label='Example curve', linestyle='dotted')
plt.annotate(r'$\theta = 2$', xy=(0.5, 0.5), xytext=(0.6, 0.5), arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.xlabel('Aggressiveness Level, r')
plt.ylabel('Target Price, τ')
plt.title("Aggressiveness Model")
plt.legend()
plt.show()
```
